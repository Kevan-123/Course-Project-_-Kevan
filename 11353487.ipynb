```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = "red_wine.csv"
data = pd.read_csv(dataset)
```

```python
data.head(10)
```

```python
data.tail(20)
```

```python
data.info()
```

```python
data.describe()
```

```python
plt.figure(figsize=(12, 10))
data['alcohol'].plot(kind = 'hist', bins=25, color = 'yellow', edgecolor = 'blue')
plt.title('Alcohol Content in Red Wine')
plt.xlabel('Alcohol Content')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['pH'], data['citric acid'], alpha = 0.5, color = 'green')
plt.title('Scatter Plot: pH vs Citric Acid')
plt.xlabel('pH')
plt.ylabel('Citric Acid Content')
plt.grid(True)
plt.show()
```

```python
quality_counts = data['quality'].value_counts()

plt.figure(figsize=(12, 10))
plt.pie(quality_counts, labels=quality_counts.index, autopct='%1.1f%%', colors=plt.cm.Set2.colors, startangle=90)

plt.title('Pie Chart: Distribution of Wine Quality')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(data['quality'], data['alcohol'], color='skyblue')
plt.title('Bar Chart: Distribution of Alcohol Content across Quality Ratings')
plt.xlabel('Quality Rating')
plt.ylabel('Alcohol Content')
plt.grid(axis='y')
plt.show()
```

```python
quality_counts = data['quality'].value_counts()

plt.figure(figsize=(8, 8))

plt.pie(quality_counts, labels=quality_counts.index, autopct='%1.1f%%', colors=plt.cm.Set3.colors, startangle=90)

centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Wine Quality Distribution')
plt.show()
```

## Histogram
The histogram shows us how the alcohol content is distributed among the red wines we have sampled in our dataset.
It serves as a fundamental tool for understanding the occurrence of various alcohol levels within the dataset
and lets us spot any trends or patterns that might be present.

### Observation
1.The histogram reveals an unbalanced distribution, with a notable concentration of the majority of values at the lower end. This suggests that a significant proportion of red wines in the dataset have lower alcohol content, resulting in a skewed pattern towards the lower range of alcohol concentrations.

2.The mode of the histogram suggests that a significant number of red wines in the dataset possess an alcohol content around 9.5.

3.The horizontal axis of the histogram shows the range of alcohol content values, helping us see how the data is spread out. We notice wines with different alcohol levels, adding to the overall variety in the dataset. This variability indicates that there is a mix of red wines with different alcohol levels.


## Scatter Plot
The scatter plot visually examines how pH levels and citric acid content are related within the red wine dataset.

### Observation
1.The points on the scatter plot are dispersed across the graph, indicating variability in both pH and citric acid levels.

2.Several points are concentrated in particular areas of the plot, suggesting the existence of patterns within the dataset.

3.The plot displays outliers, which are points that stand out from the rest. They don't follow the usual pattern of the data.


## Pie Chart
The pie chart provides a clear representation of the distribution of wine quality within the dataset.

### Observation

## Bar Chart
The bar chart effectively illustrates how alcohol content is distributed among different quality ratings in the dataset.
